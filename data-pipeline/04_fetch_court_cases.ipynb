```python
# %% [markdown]
# # 3. Court Cases Tracker
# # Source: Manual Verification + Public Records
# # Output: court_cases.json

import json
from datetime import datetime
from pydantic import BaseModel, Field
from typing import Literal

# --- Schema ---
class CourtCase(BaseModel):
    case_id: str
    defendant: str
    category: Literal["Financiación", "Difamación", "Acoso", "Fraude", "Otro"]
    status: Literal["Abierto", "Sobreseído", "Condena", "Absolución", "Pendiente"]
    court: str
    filing_date: str
    source_url: str
    is_alvise_related: bool
    credibility_badge: Literal["OFICIAL", "MEDIA", "NO_VERIFICADO"]

# --- Input Data (Manual Curation) ---
# In production, this would be a CSV you update monthly
RAW_CASES = [
    {
        "case_id": "TS-2024-001",
        "defendant": "Alvise Pérez",
        "category": "Financiación",
        "status": "Pendiente",
        "court": "Tribunal Supremo",
        "filing_date": "2024-01-15",
        "source_url": "https://www.poderjudicial.es",
        "is_alvise_related": True,
        "credibility_badge": "OFICIAL"
    },
    {
        "case_id": "TS-2024-002",
        "defendant": "Alvise Pérez",
        "category": "Difamación",
        "status": "Abierto",
        "court": "Audiencia Nacional",
        "filing_date": "2024-03-10",
        "source_url": "https://www.poderjudicial.es",
        "is_alvise_related": True,
        "credibility_badge": "OFICIAL"
    }
]

# --- Process ---
def process_cases(raw_cases):
    records = []
    for item in raw_cases:
        records.append(CourtCase(**item))
    return records

# --- Export ---
def export_data(records):
    import os
    os.makedirs("data/processed", exist_ok=True)
    OUTPUT_FILE = "data/processed/court_cases.json"
    
    with open(OUTPUT_FILE, 'w', encoding='utf-8') as f:
        json.dump({
            "meta": {
                "source": "Manual Verification / Poder Judicial",
                "generated_at": datetime.now().isoformat(),
                "disclaimer": "Data reflects public records only. Not legal advice."
            },
            "data": [r.model_dump() for r in records]
        }, f, indent=2, ensure_ascii=False)
    print(f"✅ Exported {len(records)} court cases to {OUTPUT_FILE}")

# --- Run ---
if __name__ == "__main__":
    clean = process_cases(RAW_CASES)
    export_data(clean)
```
