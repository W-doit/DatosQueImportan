```python
# %% [markdown]
# # 4. Final Consolidation & Validation
# # Goal: Ensure data is ready for Next.js /public/data folder

import json
import os
import shutil
from pathlib import Path

# --- Config ---
SOURCE_DIR = "data/processed"
DEST_DIR = "../your-nextjs-app/public/data" # Adjust path to your Lovable project folder

FILES = [
    "ine_demographics.json",
    "subsidies_chiringuitos.json",
    "court_cases.json"
]

# --- Validate ---
def validate_json(filename):
    path = os.path.join(SOURCE_DIR, filename)
    if not os.path.exists(path):
        print(f"❌ Missing: {filename}")
        return False
    try:
        with open(path, 'r') as f:
            data = json.load(f)
        if 'data' not in data or 'meta' not in data:
            print(f"❌ Invalid Structure: {filename}")
            return False
        print(f"✅ Valid: {filename} ({len(data['data'])} records)")
        return True
    except Exception as e:
        print(f"❌ Error reading {filename}: {e}")
        return False

# --- Copy ---
def deploy_data():
    if not os.path.exists(DEST_DIR):
        os.makedirs(DEST_DIR)
        
    for file in FILES:
        src = os.path.join(SOURCE_DIR, file)
        dst = os.path.join(DEST_DIR, file)
        if os.path.exists(src):
            shutil.copy(src, dst)
            print(f"🚀 Copied {file} to Next.js public folder")

# --- Run ---
if __name__ == "__main__":
    print("🔍 Validating Data...")
    all_valid = all([validate_json(f) for f in FILES])
    
    if all_valid:
        print("\n🚀 Deploying to Next.js...")
        deploy_data()
        print("\n✅ Pipeline Complete! Ready to run 'npm run dev' in your Next.js app.")
    else:
        print("\n⚠️ Fix errors before deploying.")
```
